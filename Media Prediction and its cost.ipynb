```python
# Initialize the camera
cap = cv2.VideoCapture(0)  # 0 for the default camera, you can try different numbers if you have multiple cameras

# Check if the camera is opened successfully
if not cap.isOpened():
    print("Failed to open the camera")
    exit()

# Read and display the video feed
while True:
    ret, frame = cap.read()

    if ret:
        cv2.imshow("Mobile Camera", frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) == ord('q'):
        break

# Release the camera and close the window
cap.release()
cv2.destroyAllWindows()
```

```python
#access image of mobile
import cv2
import urllib.request

# Set the URL of the IP Webcam server
url = 'http://<your_ip_address>:<port>/video'

# Open the video stream
stream = urllib.request.urlopen(url)

# Initialize the camera
cap = cv2.VideoCapture()

# Check if the camera is opened successfully
if not cap.open(stream):
    print("Failed to open the camera")
    exit()

# Read and display the video feed
while True:
    ret, frame = cap.read()

    if ret:
        cv2.imshow("Mobile Camera", frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) == ord('q'):
        break

# Release the camera and close the window
cap.release()
cv2.destroyAllWindows()
```

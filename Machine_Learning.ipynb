```python
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler

print("Using Neural Network Regression")
# Load data
path = r'C:\Users\anas_\OneDrive\Python_Backup/melb_data.csv'
data = pd.read_csv(path)
data_features = [ 'Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']

# Separate the target variable
y = data.Price

# Prepare the data
scaler = StandardScaler()
X = scaler.fit_transform(data[data_features])

# Create the model
model = MLPRegressor(hidden_layer_sizes=(100,), activation='relu', solver='adam', random_state=1, max_iter=1000)

# Train the model
model.fit(X, y)

# Make predictions
head_data = X[:5]
head_prediction = model.predict(head_data)
tail_data = X[-5:]
tail_prediction = model.predict(tail_data)
```
